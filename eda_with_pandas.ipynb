## Importortação de Bibliotecas

```python
import pandas as pd
import seaborn as sns
```

## Carregando e verificando dados

```python
df = pd.read_csv("/content/ultra_marathon_data/TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

## Limpeza de dados

```python
# Apenas corridas de 50 milhas
df[df['Event distance/length'] == '50mi']
```

```python
# Corridas de 50 milhas ou 50 km que ocorreram em 2020
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
# Eventos que ocorreram nos EUA
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# Combinação de todos os filtros
filtered_df = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
filtered_df.head()
```

```python
filtered_df.shape
```

```python
# Remoção de (USA) do nome do evento
filtered_df['Event name'].str.split('(').str.get(0)
```

```python
filtered_df['Event name'] = filtered_df['Event name'].str.split('(').str.get(0)
```

```python
filtered_df.head()
```

```python
# Limpando idade dos atletas
filtered_df['athlete_age'] = 2020 - filtered_df['Athlete year of birth']
```

```python
# Removendo h do campo Athlete performance
filtered_df['Athlete performance'] = filtered_df['Athlete performance'].str.split(' ').str.get(0)
```

```python
filtered_df.head()
```

```python
# Remover as colunas: Athlete club, Athlete country, Athlete year of birth, Athlete age category
filtered_df = filtered_df.drop(['Athlete club', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
filtered_df.head()
```

```python
# Limpeza de valores nulos
filtered_df.isna().sum()
```

```python
filtered_df[filtered_df['athlete_age'].isna()==1]
```

```python
filtered_df = filtered_df.dropna()
```

```python
filtered_df.shape
```

```python
# Valores duplicados (dupes)

filtered_df[filtered_df.duplicated() == True]
```

```python
# Reset index
filtered_df.reset_index(drop=True)
```

```python
# Fix types
filtered_df.dtypes
```

```python
filtered_df['athlete_age'] = filtered_df['athlete_age'].astype(int)
```

```python
filtered_df['Athlete average speed'] = filtered_df['Athlete average speed'].astype(float)
```

```python
filtered_df.head()
```

```python
# Renomeando colunas

filtered_df = filtered_df.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete country': 'athlete_country',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id'
})
```

```python
filtered_df.head()
```

```python
# Reordenando colunas
filtered_df = filtered_df[['race_name','race_day','race_length',
                           'race_number_of_finishers','athlete_id',
                           'athlete_gender','athlete_age',
                           'athlete_performance','athlete_average_speed',
                           'year']]
```

```python
filtered_df.head()
```

## Representação Gráfica

```python
# tabelas e gráficos

sns.histplot(filtered_df['race_length'])
```

```python
sns.histplot(
    data=filtered_df,
    x="race_length",
    hue="athlete_gender"
)
```

```python
sns.displot(filtered_df[filtered_df['race_length']=='50mi']['athlete_average_speed'])
```

```python
sns.violinplot(
    data=filtered_df,
    x='race_length',
    y='athlete_average_speed',
    hue='athlete_gender',
    split=True,
    gap=.1,
    inner="quart",
    palette="Set3"
)
```

```python
sns.lmplot(data=filtered_df, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
```

```python
# Diferença de velocidade para 50k e 50mi feminino e masculino

filtered_df.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# Média de velocidade por idade para 50mi
filtered_df.query('race_length == "50mi"')\
          .groupby('athlete_age')['athlete_average_speed']\
          .agg(['mean', 'count'])\
          .query('count > 19')\
          .sort_values(by='mean', ascending=False)
```

```python
# Análise por estação

# Spring (3-5)
# Summer (6-8)
# Fall (9-11)
# Winter (12-2)

filtered_df['race_month'] = filtered_df['race_day'].str.split('.').str.get(1).astype(int)
```

```python
filtered_df['reace_season'] = filtered_df['race_month'].apply(lambda x: 'Spring' if 3 <= x <= 5 else 'Summer' if 6 <= x <= 8 else 'Fall' if 9 <= x <= 11 else 'Winter')
```

```python
filtered_df.groupby('reace_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
```

```python
# Para 50 milhas

filtered_df.query('race_length == "50mi"').groupby('reace_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
```

```python
sns.lmplot(data=filtered_df, x='athlete_age', y='athlete_average_speed',
           col='reace_season')
```
